## Principal Component Analysis

```python
from sklearn.datasets import load_breast_cancer
```

```python
breast = load_breast_cancer()
breast_data = breast.data
breast_data.shape
```

```python
breast_label = breast.target
breast_label.shape
```

```python
import numpy as np
labels = np.reshape(breast_label,(569,1))
```

```python
final_breast_data = np.concatenate([breast_data, labels], axis=1)
final_breast_data.shape
```

```python
import pandas as pd
breast_dataset = pd.DataFrame(final_breast_data)
```

```python
features = breast.feature_names
features
```

```python
features_labels = np.append(features,'label')
```

```python
breast_dataset.columns = features_labels
breast_dataset.head()
```

```python
breast_dataset['label'].replace(0, 'Benign', inplace=True)
breast_dataset['label'].replace(1, 'Malignant', inplace=True)
breast_dataset.tail()
```

## Data Visualization using PCA

```python
from sklearn.preprocessing import StandardScaler
x = breast_dataset.loc[:, features].values
x = StandardScaler().fit_transform(x)
```

```python
x.shape
```

```python
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
normalised_breast = pd.DataFrame(x, columns=feat_cols)
normalised_breast.tail()
```

```python
from sklearn.decomposition import PCA
pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(x)
```

```python
principal_breast_DF = pd.DataFrame(data=principalComponents_breast, columns=['principal component 1', 'principal component 2'])
principal_breast_DF.tail()
```

```python
print('Explained variation per principal component: {}'.format(pca_breast.explained_variance_ratio_))
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure()
plt.figure(figsize=(10, 10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1', fontsize=20)
plt.ylabel('Principal Component - 2', fontsize=20)
plt.title('Principal Component Analysis of Breast Cancer Dataset', fontsize=20)
targets =['Benign', 'Malignant']
colors = ['r','g']
for target, color in zip(targets, colors):
    indicesToKeep = breast_dataset['label'] == target
    plt.scatter(principal_breast_DF.loc[indicesToKeep, 'principal component 1'],
                principal_breast_DF.loc[indicesToKeep, 'principal component 2'], c=color, s=50)

plt.legend(targets,prop={'size': 15})
plt.show()
```

## K-Means with Python

```python
from sklearn.cluster import KMeans
```

```python
X = -2 * np.random.rand(100, 2)
X1 = 1 + 2 *np.random.rand(50, 2)
X[50:100, :] = X1

plt.scatter(X[ :, 0], X[ :, 1], s= 50, c='b')
plt.show()
```

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') #within cluster sum of squares
plt.show()
```

```python
Kmean = KMeans(n_clusters=2)
Kmean.fit(X)
```

```python
Kmean.cluster_centers_
```

```python
plt.scatter(X[ : , 0], X[ :, 1], s=50, c='b')
plt.scatter(-0.95436043, -1.01560962, s=200, c='g', marker='s')
plt.scatter(2.06644818,  2.08460163, s=200, c='r', marker='s')
plt.show()
```

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(
    n_samples=150, n_features=2,
    centers=3, cluster_std=0.5,
    shuffle=True, random_state=0
)

plt.scatter(
    X[:, 0], X[:, 1],
    c='white', marker='o',
    edgecolor='black', s=50
)
plt.show()
```

```python
km = KMeans(n_clusters=3, init='random', max_iter=300)
y_km = km.fit_predict(X)
```

```python
plt.scatter(
    X[y_km == 0, 0], X[y_km == 0, 1],
    s=50, c='lightgreen',
    marker='s', edgecolor='black',
    label='cluster 1'
)

plt.scatter(
    X[y_km == 1, 0], X[y_km == 1, 1],
    s=50, c='orange',
    marker='o', edgecolor='black',
    label='cluster 2'
)

plt.scatter(
    X[y_km == 2, 0], X[y_km == 2, 1],
    s=50, c='lightblue',
    marker='v', edgecolor='black',
    label='cluster 3'
)

# plot the centroids
plt.scatter(
    km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],
    s=250, marker='*',
    c='red', edgecolor='black',
    label='centroids'
)
plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

```python
distortions = []
for i in range(1, 11):
    km = KMeans(
        n_clusters=i, init='random',
        n_init=10, max_iter=300,
        tol=1e-04, random_state=0
    )
    km.fit(X)
    distortions.append(km.inertia_)

# plot
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

## Save and Load Models

```python
weather=['Sunny','Sunny','Overcast','Rainy','Rainy','Rainy','Overcast','Sunny','Sunny', 'Rainy','Sunny','Overcast','Overcast','Rainy']
temp=['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild','Cool','Mild','Mild','Mild','Hot','Mild']
play=['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']

from sklearn import preprocessing
le = preprocessing.LabelEncoder()
weather_encoded=le.fit_transform(weather)
temp_encoded=le.fit_transform(temp)
label=le.fit_transform(play)


features=list(zip(weather_encoded,temp_encoded))
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(features,label)

import pickle
filename = 'finalized_model.sav'
pickle.dump(model, open(filename, 'wb'))
```

```python
loaded_model = pickle.load(open('finalized_model.sav', 'rb'))
loaded_model.predict([[0,2]])
```
